```python
# Q1: Estimation Statistics
"""
- Estimation statistics involves using sample data to infer or estimate population parameters.
- Point Estimate: A single value estimate for a parameter. Example: Sample mean as an estimate of population mean.
- Interval Estimate: A range of values within which the parameter is likely to fall. Example: Confidence interval.
"""

# Q2: Python function to estimate population mean
def estimate_population_mean(sample_mean, sample_std, sample_size):
    from math import sqrt
    margin_of_error = 1.96 * (sample_std / sqrt(sample_size))  # 95% confidence level
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error
    return lower_bound, upper_bound

# Q3: Hypothesis Testing
"""
- Hypothesis testing is a statistical method to determine whether there is enough evidence to reject a null hypothesis.
- Importance:
  1. Validates claims or assumptions about a population parameter.
  2. Helps in making informed decisions using data.
"""

# Q4: Hypothesis Example
"""
H0: The average weight of male college students is equal to the average weight of female college students.
H1: The average weight of male college students is greater than the average weight of female college students.
"""

# Q5: Hypothesis test for two population means
import scipy.stats as stats

def two_sample_t_test(sample1, sample2):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    return t_stat, p_value

# Q6: Null and Alternative Hypothesis
"""
- Null Hypothesis (H0): Assumes no effect or no difference. Example: The mean revenue of two stores is equal.
- Alternative Hypothesis (H1): Assumes an effect or difference. Example: The mean revenue of two stores is different.
"""

# Q7: Steps in Hypothesis Testing
"""
1. Define null and alternative hypotheses.
2. Select a significance level (alpha).
3. Choose an appropriate test statistic.
4. Calculate the test statistic and p-value.
5. Compare p-value with alpha to accept or reject H0.
"""

# Q8: p-Value
"""
- Definition: The probability of observing test results at least as extreme as the results observed, under the null hypothesis.
- Significance: Helps decide whether to reject H0. A smaller p-value indicates stronger evidence against H0.
"""

# Q9: Student's t-distribution plot
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=10)
plt.plot(x, y, label="t-distribution (df=10)")
plt.title("Student's t-Distribution")
plt.xlabel("t")
plt.ylabel("Density")
plt.legend()
plt.show()

# Q10: Two-sample t-test for independent samples
def independent_t_test(sample1, sample2):
    t_stat, p_value = stats.ttest_ind(sample1, sample2)
    return t_stat, p_value

# Q11: Student’s t-distribution
"""
- Definition: A probability distribution used for small sample sizes or when population variance is unknown.
- Usage: Used in t-tests for hypothesis testing.
"""

# Q12: t-Statistic
"""
Formula: t = (X̄ - μ) / (s / sqrt(n))
Where:
X̄ = Sample mean
μ = Population mean
s = Sample standard deviation
n = Sample size
"""

# Q13: Confidence interval for coffee shop revenue
sample_mean = 500
sample_std = 50
sample_size = 50
confidence_interval = estimate_population_mean(sample_mean, sample_std, sample_size)
"""
95% Confidence Interval: (confidence_interval)
"""

# Q14: Hypothesis test for blood pressure
sample_mean = 8
sample_std = 3
sample_size = 100
hypothesized_mean = 10
t_stat = (sample_mean - hypothesized_mean) / (sample_std / np.sqrt(sample_size))
p_value = stats.t.sf(np.abs(t_stat), df=sample_size - 1) * 2
"""
t-statistic: (t_stat), p-value: (p_value)
"""

# Q15: Hypothesis test for product weight
sample_mean = 4.8
sample_std = 0.5
sample_size = 25
hypothesized_mean = 5
t_stat = (sample_mean - hypothesized_mean) / (sample_std / np.sqrt(sample_size))
p_value = stats.t.sf(np.abs(t_stat), df=sample_size - 1)
"""
t-statistic: (t_stat), p-value: (p_value)
"""

# Q16: Two-sample t-test for student groups
mean1, std1, n1 = 80, 10, 30
mean2, std2, n2 = 75, 8, 40
t_stat, p_value = stats.ttest_ind_from_stats(mean1, std1, n1, mean2, std2, n2)
"""
t-statistic: (t_stat), p-value: (p_value)
"""

# Q17: Confidence interval for ad viewers
sample_mean = 4
sample_std = 1.5
sample_size = 50
confidence_interval = estimate_population_mean(sample_mean, sample_std, sample_size)
"""
99% Confidence Interval: (confidence_interval)
"""
```
