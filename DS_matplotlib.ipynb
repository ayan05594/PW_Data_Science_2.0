```python
import matplotlib.pyplot as plt
import numpy as np

# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
# Matplotlib is a Python library used for creating static, animated, and interactive visualizations in Python.
# It is used for plotting data in the form of graphs, charts, and plots.

# Five common types of plots in Matplotlib:
# 1. Line plot
# 2. Scatter plot
# 3. Bar plot
# 4. Histogram
# 5. Box plot


# Q2: What is a scatter plot? Use the following code to generate data for x and y.
# Plot a scatter plot.
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()


# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

# The subplot() function is used in Matplotlib to create multiple subplots within a single figure.
# It allows you to organize multiple plots in a grid layout.

# Data for the four line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Plotting four subplots
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)  # 2x2 grid, first subplot
plt.plot(x1, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)  # 2x2 grid, second subplot
plt.plot(x2, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)  # 2x2 grid, third subplot
plt.plot(x3, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)  # 2x2 grid, fourth subplot
plt.plot(x4, y4)
plt.title("Line 4")

plt.tight_layout()  # Adjust spacing to prevent overlap
plt.show()


# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# A bar plot is used to represent categorical data with rectangular bars where the length of the bar is proportional to the value.
# A bar plot can be either vertical or horizontal.

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar plot
plt.bar(company, profit)
plt.title("Bar Plot: Company Profit")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title("Horizontal Bar Plot: Company Profit")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()


# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# A box plot (or box-and-whisker plot) is used to represent the distribution of a dataset. It shows the median, quartiles, and outliers.
# The plot typically includes a box, whiskers, and dots representing outliers.

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Box plot for two datasets
plt.boxplot([box1, box2], labels=['Box1', 'Box2'])
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```
