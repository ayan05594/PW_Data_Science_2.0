```python
# Q1. What is multithreading in python? Why is it used? Name the module used to handle threads in python.

# Multithreading is a concurrent execution technique where different parts of a program are executed 
# simultaneously in separate threads. It's used to perform multiple tasks concurrently, 
# which can lead to improved performance in programs that have tasks like I/O-bound or network-bound processes. 

# The module used to handle threads in Python is the 'threading' module.


# Q2. Why threading module used? Write the use of the following functions:

# The 'threading' module in Python is used for creating and managing threads to execute 
# different tasks concurrently. It provides an easy way to implement multithreading.

# ActiveCount() - This function returns the number of Thread objects currently alive.
# CurrentThread() - This function returns the current Thread object corresponding to the callerâ€™s thread of execution.
# Enumerate() - This function returns a list of all Thread objects currently alive.


# Q3. Explain the following functions:

# run() - The 'run()' method is used to define the task to be executed in a thread.
# It is called when the thread is started.

# start() - The 'start()' method is used to begin the thread's activity.
# It invokes the run() method of the thread.

# join() - The 'join()' method blocks the execution of the calling thread until the thread whose join() method was called has finished executing.

# isAlive() - The 'isAlive()' method checks whether a thread is still executing or not.


# Q4. Write a python program to create two threads. Thread one must print the list of squares and thread two must print the list of cubes.

import threading

# Function to print squares
def print_squares():
    squares = [i ** 2 for i in range(1, 6)]
    print("Squares:", squares)

# Function to print cubes
def print_cubes():
    cubes = [i ** 3 for i in range(1, 6)]
    print("Cubes:", cubes)

# Create two threads
thread1 = threading.Thread(target=print_squares)
thread2 = threading.Thread(target=print_cubes)

# Start the threads
thread1.start()
thread2.start()

# Wait for both threads to finish
thread1.join()
thread2.join()

print("Both threads have finished executing.")


# Q5. State advantages and disadvantages of multithreading.

# Advantages:
# 1. Improved application performance through parallel execution of tasks.
# 2. Multithreading makes efficient use of CPU when dealing with I/O-bound operations.
# 3. Enhances user experience by allowing the application to remain responsive.

# Disadvantages:
# 1. Increased complexity in code design and debugging.
# 2. Issues like race conditions and deadlocks can occur if threads aren't managed properly.
# 3. Multithreading doesn't always lead to performance gains for CPU-bound tasks.


# Q6. Explain deadlocks and race conditions.

# Deadlocks occur when two or more threads are blocked forever, waiting for each other to release resources. 
# This can happen when each thread holds a resource that the other thread needs to proceed.

# Race conditions occur when multiple threads access shared data or resources concurrently and the final outcome 
# depends on the sequence or timing of the threads' execution, leading to unpredictable results.
```
