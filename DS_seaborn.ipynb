```python
import seaborn as sns
import matplotlib.pyplot as plt

# Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
# 1. **Line Plot**: Useful to visualize the relationship between a continuous variable and time or any other variable.
# 2. **Box Plot**: Used to show the distribution of a dataset through its quartiles and can highlight outliers.
# 3. **Heatmap**: Used to visualize matrix-style data, often used for correlation matrices or pivot tables.
# 4. **Histogram**: Displays the distribution of a single continuous variable, showing the frequency of data points in bins.
# 5. **Pair Plot**: Useful for exploring relationships between multiple continuous variables in a dataset.

# Q2: Load the "fmri" dataset using the load_dataset function of seaborn.
fmri = sns.load_dataset("fmri")

# Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
plt.title("FMRI Line Plot")
plt.show()


# Q3: Load the "titanic" dataset using the load_dataset function of seaborn.
titanic = sns.load_dataset("titanic")

# Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.
plt.figure(figsize=(10, 5))

# Box plot for age by pclass
plt.subplot(1, 2, 1)
sns.boxplot(x="pclass", y="age", data=titanic)
plt.title("Box Plot of Age by Pclass")

# Box plot for fare by pclass
plt.subplot(1, 2, 2)
sns.boxplot(x="pclass", y="fare", data=titanic)
plt.title("Box Plot of Fare by Pclass")

plt.tight_layout()
plt.show()


# Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
diamonds = sns.load_dataset("diamonds")

# Plot histogram for price column with hue parameter for cut
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack")
plt.title("Histogram of Diamond Prices by Cut")
plt.show()


# Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
iris = sns.load_dataset("iris")

# Plot pair plot with hue parameter for species
sns.pairplot(iris, hue="species")
plt.title("Pair Plot of Iris Dataset")
plt.show()


# Q6: Use the "flights" dataset from seaborn to plot a heatmap.
flights = sns.load_dataset("flights")

# Pivot the data for the heatmap
flights_pivot = flights.pivot_table(index="month", columns="year", values="passengers")

# Plot heatmap with appropriate format for float values
sns.heatmap(flights_pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Heatmap of Flight Passengers")
plt.show()
```
