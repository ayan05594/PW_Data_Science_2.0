```python
import pandas as pd
import numpy as np

# Data for course_name and duration
course_name = ['Data Science', 'Machine Learning', 'Big Data', 'Data Engineer']
duration = [2, 3, 6, 4]
df = pd.DataFrame(data={'course_name': course_name, 'duration': duration})

# Q1. Print the data present in the second row of the dataframe
print("Q1. Data in the second row of the dataframe:")
print(df.iloc[1])

# Q2. Difference between loc and iloc
print("\nQ2. Difference between loc and iloc:")
# loc is label-based, iloc is position-based
print("loc: Selects rows and columns by label names")
print("iloc: Selects rows and columns by integer position (0-indexed)")

# Q3. Reindex the DataFrame and display the results
print("\nQ3. Reindex the DataFrame using reindex = [3,0,1,2]:")
new_df = df.reindex([3, 0, 1, 2])
print(new_df)

# Access by loc and iloc after reindexing
print("\nnew_df.loc[2]:")
print(new_df.loc[2])  # Loc-based access

print("\nnew_df.iloc[2]:")
print(new_df.iloc[2])  # Iloc-based access

# Q4. Statistical measurements
print("\nQ4. Statistical measurements:")

# Q4(i). Mean of each column
print("\nMean of each column:")
print(df.select_dtypes(include=[np.number]).mean())  # mean only for numeric columns

# Q4(ii). Standard deviation of 'duration' column
print("\nStandard deviation of 'duration' column:")
print(df['duration'].std())

# Q5. Replace the data present in the second row of column 'duration' by a string variable
print("\nQ5. Replacing the second row of column 'duration' with a string:")
df.loc[1, 'duration'] = 'string_value'  # Replacing with a string
print(df)
# Recalculate mean after replacing with a string (will cause an error)
try:
    print("\nMean of 'duration' after replacement:")
    print(df['duration'].mean())
except Exception as e:
    print(f"Error: {e}")

# Q6. Windows function in pandas
print("\nQ6. Windows function in pandas:")
# A window function allows you to perform calculations over a specified range of values
# Types of window functions: rolling, expanding, and exponentially weighted

# Q7. Print the current month and year
print("\nQ7. Current month and year:")
print(pd.to_datetime('today').strftime('%Y-%m'))

# Q8. Calculate the difference between two dates in days, hours, and minutes
from datetime import datetime
print("\nQ8. Calculate the difference between two dates:")
date1 = input("Enter the first date (YYYY-MM-DD): ")
date2 = input("Enter the second date (YYYY-MM-DD): ")

date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)
delta = date2 - date1

days = delta.days
hours = delta.seconds // 3600
minutes = (delta.seconds // 60) % 60

print(f"Difference: {days} days, {hours} hours, {minutes} minutes")

# Q9. Convert specified column to categorical data type
print("\nQ9. Convert column to categorical data type:")
file_path = input("Enter the file path of the CSV file: ")
df = pd.read_csv(file_path)

column_name = input("Enter the column name to convert to categorical: ")
df[column_name] = pd.Categorical(df[column_name])

category_order = input("Enter the category order (comma separated): ").split(',')
df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

# Display the sorted data
print(df.sort_values(by=column_name))

# Q10. Stacked bar chart visualization
import matplotlib.pyplot as plt
print("\nQ10. Stacked bar chart:")
file_path = input("Enter the file path of the sales CSV file: ")
df_sales = pd.read_csv(file_path)

df_sales.set_index('Time').T.plot(kind='bar', stacked=True)
plt.show()

# Q11. Mean, median, mode of test scores
print("\nQ11. Mean, median, mode of test scores:")
file_path = input("Enter the file path of the student data CSV: ")
df_student = pd.read_csv(file_path)

mean_score = df_student['Test Score'].mean()
median_score = df_student['Test Score'].median()
mode_score = df_student['Test Score'].mode()

# Display the results in a table
print(f"\n+-----------+--------+")
print(f"| Statistic | Value |")
print(f"+-----------+--------+")
print(f"| Mean      | {mean_score} |")
print(f"| Median    | {median_score} |")
print(f"| Mode      | {', '.join(map(str, mode_score))} |")
print(f"+-----------+--------+")
```
