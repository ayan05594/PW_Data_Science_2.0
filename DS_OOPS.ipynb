```python
# Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.
# A class is a blueprint for creating objects. It defines a set of attributes and methods that the objects created from the class will have.
# An object is an instance of a class. It contains data and methods that operate on the data.

# Example:
class Car:
    # Class attributes
    wheels = 4

    # Constructor method (__init__) to initialize object
    def __init__(self, make, model, color):
        self.make = make
        self.model = model
        self.color = color

    # Method to display car details
    def display_info(self):
        print(f"This is a {self.color} {self.make} {self.model} with {self.wheels} wheels.")

# Create an object of the Car class
car1 = Car("Toyota", "Corolla", "Red")
car1.display_info()  # Output: This is a Red Toyota Corolla with 4 wheels.


# Q2. Name the four pillars of OOPs.
# The four pillars of Object-Oriented Programming (OOP) are:
# 1. Encapsulation: Bundling data and methods that operate on the data into a single unit (class).
# 2. Abstraction: Hiding the internal implementation details and showing only the necessary features to the user.
# 3. Inheritance: A mechanism that allows one class to inherit attributes and methods from another class.
# 4. Polymorphism: The ability of a function or method to work with different types of data or to override methods in derived classes.


# Q3. Explain why the __init__() function is used. Give a suitable example.
# The __init__() function is a special method in Python classes. It is automatically called when an object of the class is created.
# The purpose of __init__() is to initialize the object's attributes.

# Example:
class Dog:
    # Constructor to initialize object
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    def speak(self):
        print(f"{self.name} says Woof!")

# Create an object of the Dog class
dog1 = Dog("Buddy", "Golden Retriever")
dog1.speak()  # Output: Buddy says Woof!


# Q4. Why self is used in OOPs?
# The 'self' keyword is used in Python to represent the instance of the class. It allows access to the attributes and methods of the class.
# It is required to reference the instance of the object being created and to differentiate between instance variables and local variables.

# Example:
class Person:
    def __init__(self, name, age):
        self.name = name  # 'self.name' is the instance attribute
        self.age = age    # 'self.age' is the instance attribute

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Create an object of the Person class
person1 = Person("Alice", 30)
person1.greet()  # Output: Hello, my name is Alice and I am 30 years old.


# Q5. What is inheritance? Give an example for each type of inheritance.
# Inheritance is a mechanism in object-oriented programming where a class (child class) can inherit attributes and methods from another class (parent class).
# Types of inheritance:
# 1. Single Inheritance: A subclass inherits from one superclass.
# 2. Multiple Inheritance: A subclass inherits from multiple superclasses.
# 3. Multilevel Inheritance: A subclass inherits from another subclass.
# 4. Hierarchical Inheritance: Multiple subclasses inherit from a single superclass.

# Example of Single Inheritance
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):  # Dog inherits from Animal
    def bark(self):
        print("Dog barks")

dog1 = Dog()
dog1.speak()  # Inherited from Animal class
dog1.bark()   # Defined in Dog class

# Example of Multiple Inheritance
class Bird:
    def fly(self):
        print("Bird can fly")

class Bat(Animal, Bird):  # Bat inherits from both Animal and Bird
    def hang(self):
        print("Bat hangs upside down")

bat1 = Bat()
bat1.speak()  # Inherited from Animal class
bat1.fly()    # Inherited from Bird class
bat1.hang()   # Defined in Bat class

# Example of Multilevel Inheritance
class Vehicle:
    def start(self):
        print("Vehicle starts")

class Car(Vehicle):  # Car inherits from Vehicle
    def drive(self):
        print("Car is driving")

class ElectricCar(Car):  # ElectricCar inherits from Car (which inherits from Vehicle)
    def charge(self):
        print("Electric car is charging")

ev1 = ElectricCar()
ev1.start()  # Inherited from Vehicle class
ev1.drive()  # Inherited from Car class
ev1.charge() # Defined in ElectricCar class

# Example of Hierarchical Inheritance
class Shape:
    def draw(self):
        print("Drawing a shape")

class Circle(Shape):  # Circle inherits from Shape
    def draw(self):
        print("Drawing a Circle")

class Square(Shape):  # Square inherits from Shape
    def draw(self):
        print("Drawing a Square")

circle1 = Circle()
circle1.draw()  # Output: Drawing a Circle

square1 = Square()
square1.draw()  # Output: Drawing a Square
```
