```python
# Q1. Create one variable containing the following type of data:
# (i) string
string_data = "Hello, World!"

# (ii) list
list_data = [1, 2, 3, 4, 5]

# (iii) float
float_data = 3.14

# (iv) tuple
tuple_data = (1, 2, 3, 4, 5)

# Q2. Given are some following variables containing data:
# (i) var1 = ' '
# (ii) var2 = '[ DS , ML , Python]'
# (iii) var3 = ['DS', 'ML', 'Python']
# (iv) var4 = 1.

var1 = ' '  # String with a space, data type: str
var2 = '[ DS , ML , Python]'  # String containing a list in string form, data type: str
var3 = ['DS', 'ML', 'Python']  # List of strings, data type: list
var4 = 1.  # Float value, data type: float

# Printing data types of the variables
print(f"Data type of var1: {type(var1)}")
print(f"Data type of var2: {type(var2)}")
print(f"Data type of var3: {type(var3)}")
print(f"Data type of var4: {type(var4)}")

# Q3. Explain the use of the following operators using an example:
# (i) /
# Division operator
a = 10
b = 2
division_result = a / b  # Output: 5.0

# (ii) %
# Modulus operator - returns the remainder of division
modulus_result = a % b  # Output: 0

# (iii) //
# Floor division operator - returns the largest integer less than or equal to the result of division
floor_division_result = a // b  # Output: 5

# (iv) **
# Exponentiation operator - returns the result of raising the first number to the power of the second number
exponentiation_result = a ** b  # Output: 100

# Displaying results
print(f"Division result: {division_result}")
print(f"Modulus result: {modulus_result}")
print(f"Floor division result: {floor_division_result}")
print(f"Exponentiation result: {exponentiation_result}")

# Q4. Create a list of length 10 of your choice containing multiple types of data. Using for loop print the element and its data type.
mixed_list = [1, 'hello', 3.14, True, (1, 2), ['a', 'b'], 42, None, 'world', 10]

# Using for loop to print the element and its data type
for element in mixed_list:
    print(f"Element: {element}, Data type: {type(element)}")

# Q5. Using a while loop, verify if the number A is purely divisible by number B and if so, then how many times it can be divisible.
A = 100
B = 5
count = 0

# While loop to check divisibility
while A % B == 0:
    A = A // B
    count += 1

print(f"The number can be divided {count} times.")

# Q6. Create a list containing 25 int type data. Using for loop and if-else condition print if the element is divisible by 3 or not.
int_list = [i for i in range(1, 26)]  # List of numbers from 1 to 25

# Using for loop and if-else to check divisibility by 3
for num in int_list:
    if num % 3 == 0:
        print(f"{num} is divisible by 3")
    else:
        print(f"{num} is not divisible by 3")

# Q7. What do you understand about mutable and immutable data types? Give examples for both showing this property.
# Mutable data types: The objects can be changed or modified after they are created.
# Example: List
mutable_data = [1, 2, 3]
mutable_data[0] = 10  # Modifying the list
print(f"Modified mutable data: {mutable_data}")

# Immutable data types: The objects cannot be modified after they are created.
# Example: Tuple
immutable_data = (1, 2, 3)
# immutable_data[0] = 10  # This will raise an error because tuples are immutable
print(f"Immutable data: {immutable_data}")
```
